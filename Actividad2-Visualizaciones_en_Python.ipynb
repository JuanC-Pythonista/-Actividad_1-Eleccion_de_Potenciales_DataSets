<h3><b>Conjunto de datos lista de juegos y características plataforma Steam - (All_Steam_Games_DataSet)</b></h3>

```python
# Se importan librerías
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
#Se carga el DataSet
url = "https://media.githubusercontent.com/media/JuanC-Pythonista/Proyecto_Data_Science_Unidad1/main/DataSetBootcamp/All_Steam_Games_DataSet.csv"
DataSet = pd.read_csv(url, sep=";", encoding="latin-1")
```

```python
#Se remplaza el punto en la columna Owners por “nada” para que los valores sean mas aptos para operaciones
DataSet["Owners"] = DataSet["Owners"].str.replace(".", "")
```

```python
#Se cambia el tipo de dato de la variable Owners de Object a Int64 para que sea de tipo numérico
DataSet["Owners"] = DataSet["Owners"].astype("int64")
```

```python
#Se utiliza el set_option para aumentar la cantidad de columnas a visualizar y se muestran los valores por medio 
# de una tabla del DataSet cargado 
pd.set_option("display.max_columns", None)
DataSet.head()
```

```python
#Se revisan los tipos de datos de las variables 
DataSet.info()
```

<h5><u>Variables Potencialmente Interesantes</u></h5>
<p class="text-justify">
Objetivo: predicciones para desarrolladores de videojuegos plataforma STEAM <br><br>
Con respecto al DataSet1 (Juegos Plataforma Steam) a mi consideración las variables importantes son:<br><br>
•	<b>Developer:</b> La variable desarrollador es de importancia puesto que evidencia cuales empresas de desarrollo han tenido más éxito.<br>
•	<b>Genre:</b> La variable genero evidencia que cual es el genero mas buscado por los Gamers.<br>
•	<b>Positive Reviews:</b> La cantidad de reseñas positivas evidencian la aceptación del videojuego.<br>
•	<b>Negative Reviews:</b> La cantidad de reseñas negativas evidencian la negación al videojuego.<br>
•	<b>CCU (Concurrent Steam Users):</b> La concurrencia de los usuarios Steam frente al video juego demuestra la cantidad de aficionados que posee.<br>
•	<b>Owners:</b> La variable propietarios demuestra la cantidad de usuarios que han comprado el videojuego.
</p>

<h2><b>Visualizaciones en Python</b></h2>

```python
#Para empezar con la visualización de datos primero se deja como index la columna Developers, luego se elimina
#dicha columna de su posición original para que no se encuentre duplicidad de datos
DataSet.index = DataSet["Developer"]
DataSet = DataSet.drop("Developer", axis="columns")
```

```python
#Nuevamente se muestran los valores por medio de una tabla del DataSet cargado, ahora con el cambio del index
DataSet.head()
```

```python
#Se realiza el primer GroupBy y Sort con la variable Developer y se realiza una sumatoria para ver los valores
#unificados, por otro lado, se ordenan de manera descendente con la variable Owners y finalmente se muestra la 
#tabla agrupada y ordenada
DataSetSum0=DataSet.groupby(['Developer'],sort = False).sum()
DataSetSort0=DataSetSum0.sort_values('Owners', ascending=False)
DataSetSort0.head(5)
```

```python
#Se realiza el segundo GroupBy y Sort ahora con la variable Genre y se realiza una sumatoria para ver los valores 
#unificados, por otro lado, se ordenan de manera descendente con la variable Owners y finalmente se muestra la 
#tabla agrupada y ordenada
DataSetSum1=DataSet.groupby(['Genre'],sort = False).sum()
DataSetSort1=DataSetSum1.sort_values('Owners', ascending=False)
DataSetSort1.head(5)
```

<h2><b>Matplotlib</b></h2>

```python
# Se realiza el primero grafico por medio de la librería Matplotlib que corresponde a un diagrama de torta
fig, ax = plt.subplots(figsize=(8, 6))
ax.pie(
    DataSetSort0["Owners"].head(8),
    labels= DataSetSort0.head(8).index,
    autopct="%0.1f %%",
    wedgeprops={"edgecolor": "black"},
)
plt.title('Gráfico torta – Compradores por desarrolladores', size=12,fontweight="bold")
```

<p class="text-justify"><b>
El grafico de torta pretende mostrar el porcentaje de compradores con respecto a los desarrolladores de los videojuegos, según el grafico el desarrollador con más compradores es Valve Corporation, seguido de manera distante por KRAFTON, Inc. y Amazon Games
</p>

```python
# Se realiza el segundo grafico por medio de la librería Matplotlib, este corresponde a un diagrama de líneas
fig, (ax) = plt.subplots(
    nrows=1,
    ncols=1,
    figsize=(17, 4),
)  
ax.plot(DataSetSort0.head(10).index, DataSetSort0["Positive Reviews"].head(10), color="b")
ax.set_xlabel(r"$\mathbf{Desarrolladores}$")
ax.set_ylabel(r"$\mathbf{Reseñas}$")
plt.title('Numero de Reseñas Positivas Video Juegos por Desarrollador', size=15,fontweight="bold")
```

<p class="text-justify"><b>Diagramas de líneas pretende mostrar cual desarrollador ha recibido más reseñas positivas, según el grafico el desarrollador con mejores reseñases es Valve Corporation, seguido por valve- Hidden Path Entertainment</p>

<h2><b>Seaborn</b></h2>

```python
#Se realiza el primer grafico por medio de la librería SeaBorn, este corresponde a un histograma
sns.barplot(data=DataSetSort1.head(10), x="CCU", y=DataSetSort1.head(10).index)
plt.title('Histograma Concurrent Steam Users y géneros ', size=12,fontweight="bold")
```

<p class="text-justify"><b>El Histograma pretende mostrar la variable Concurrent Steam Users con respecto al género del video juego, según el grafico la mayor concurrencia de usuarios Steam (CCU) se da en el género Action, Free to Play seguido por el género Action, Free to Play, Strategy</p>

```python
# Se realiza el segundo grafico por medio de la librería SeaBorn, este corresponde a una dispersión
sns.scatterplot(data=DataSetSort0.head(10), x="Negative Reviews", y="Price",hue=DataSetSort0.head(10).index,alpha=1)
plt.title('Análisis bivariable numero de reseñas Negativas y Precio', size=12,fontweight="bold")
```

<p class="text-justify"><b>El grafico de dispersión pretende mostrar la relación entre las reseñas negativas y el precio del video juego, según el grafico no se tiene mucha relación entre las variables, pero se da una ligera correlación positiva pues al aumentar el precio aumentan las reseñas negativas de manera muy tenue, también se encuentran valores atípicos. </p>
