<p class="text-justify">
Objetivo: predicciones para desarrolladores de videojuegos plataforma STEAM <br><br>
Con respecto al DataSet1 (Juegos Plataforma Steam) a mi consideración las variables importantes son:<br><br>

•	<b>CCU (Concurrent Steam Users):</b> La concurrencia de los usuarios Steam frente al video juego demuestra la cantidad de aficionados que posee.<br>
•	<b>Owners:</b> La variable propietarios demuestra la cantidad de usuarios que han comprado el videojuego.

<h5><u>DataSet2: Variables Potencialmente Interesantes</u></h5>

<p class="text-justify">Objetivo: predicciones de futuros malware según permisos en el sistema<br><br>
Con respecto al DataSet2 (Permisos aplicaciones Android) a mi consideración las variables importantes son:<br><br>
•	<b>App:</b> La variable nombre de la aplicación es de importancia puesto que permitirá saber al usuario bajo que nombre se encuentra el posible malware.<br>
•	<b>Category:</b> La categoría de la App es una variable de importancia dado que se podrá acotar las categorías donde mas se desarrollan malwares y poder llevar un control.<br>
•	<b>Rating:</b> El rating es una variable de importancia para predecir la atracción de los usuarios por ciertas apps.<br>
•	<b>Price:</b> El precio de la app es una variable de importancia para saber si los malware los ocultan bajo apps gratis o de precios muy bajos.<br>
•	<b>Dangerous permissions count:</b> La variable contador de permisos peligrosos es muy importante pues evidenciara que tantos permisos solicita una App para considerarla preligrosa.<br>
•	<b>Safe permissions count:</b> La variable contador de permisos seguros es de importancia pues estos permisos no presentan ningún riesgo en el sistema</p>
